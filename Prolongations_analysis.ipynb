# Prolongation coefficients evaluations for each month and for each manager + the whole department


For each month N:


*  Coefficient 1 (prolongation in the first month after project was completed):

1.   Evaluate the shipment amount in the month of completion (N-1) - denominator_1
2.   Evaluate the shipment amount in the month N (current month) - numerator_1
3.   Coefficient_1 = numerator_1 / denominator_1

*   Coefficient 2 (prolongation in the second month after project was completed):
For projects completed in B-2 and not prolonged in N-1:

1.    Evaluate the shipment amount in the month of completion (N-2) - denominator_2
2.   Evaluate the shipment amount in the month N (current month) if the projects wasn't prolonged in  N-1 - numerator_2
3.   Coefficient_2 = numerator_2 / denominator_2




```python
all_months = sorted(merged_df['Месяц отгрузки'].unique())

monthly_res = []

for i in range(2, len(all_months)):

    current_month = all_months[i]
    prev_month = all_months[i - 1]
    prev_prev_month = all_months[i - 2]

    for manager in list(merged_df['AM'].unique()) + ['Весь отдел']:

      if manager == 'Весь отдел':
        current_manager_df = merged_df.copy()
      else:
        current_manager_df = merged_df[merged_df['AM'] == manager].copy()

      completed_last_month = current_manager_df[current_manager_df['Последний месяц реализации проекта'] == prev_month]

      shipment_in_prev_month = completed_last_month[completed_last_month['Месяц отгрузки'] == prev_month]
      denom_1 = shipment_in_prev_month['Сумма отгрузки'].sum()

      shipment_in_current_month = completed_last_month[completed_last_month['Месяц отгрузки'] == current_month]
      numer_1 = shipment_in_current_month['Сумма отгрузки'].sum()

      coeff_1 = numer_1 / denom_1 if denom_1 > 0 else np.nan

      completed_2_months_ago = current_manager_df[current_manager_df['Последний месяц реализации проекта'] == prev_prev_month]

      shipment_last_month = current_manager_df[
            (current_manager_df['Последний месяц реализации проекта'] == prev_prev_month) &
            (current_manager_df['Месяц отгрузки'] == prev_month)
            ]

      ids_shipment_last_month = shipment_last_month['id'].unique()

      no_shipment_last_month = completed_2_months_ago[~completed_2_months_ago['id'].isin(ids_shipment_last_month)]

      denom_2 = no_shipment_last_month[no_shipment_last_month['Месяц отгрузки'] == prev_prev_month]['Сумма отгрузки'].sum()

      numer_2 = no_shipment_last_month[no_shipment_last_month['Месяц отгрузки'] == current_month]['Сумма отгрузки'].sum()

      coeff_2 = numer_2 / denom_2 if denom_2 > 0 else np.nan

      monthly_res.append({
          'Месяц анализа (N)': current_month,
          'Менеджер': manager,
          'Коэффициент 1': round(coeff_1, 4) if not np.isnan(coeff_1) else np.nan,
          'Коэффициент 2': round(coeff_2, 4) if not np.isnan(coeff_2) else np.nan,
      })

monthly_results_df = pd.DataFrame(monthly_res)
```

# Aggregated prolongation coefficients evaluations for each year and for each manager + the whole department

The prolongation coefficients for each year are evaluated similarly to the same for each month, but in this case, both the denominator and numerator are summed for each month of a specific year.


```python
merged_df['Год отгрузки'] = merged_df['Месяц отгрузки'].dt.year

all_managers = list(merged_df['AM'].unique())
all_managers.append("Весь отдел")

yearly_res = []

for manager in all_managers:
    if manager == "Весь отдел":
        current_manager_df = merged_df.copy()
    else:
        current_manager_df = merged_df[merged_df["AM"] == manager].copy()

    sorted_years = sorted(current_manager_df['Год отгрузки'].unique())

    for year in sorted_years:

        if year == 2024:
          continue

        # year_df = current_manager_df[current_manager_df['Год отгрузки'] == year]

        completed = current_manager_df[current_manager_df['Последний месяц реализации проекта'].dt.year == year]
        completed_ids = completed[['id', 'Последний месяц реализации проекта']].drop_duplicates()

        denom_1 = 0
        numer_1 = 0

        for _, row in completed_ids.iterrows():
            id = row['id']
            last_month = row['Последний месяц реализации проекта']

            shipment_in_completion_month = current_manager_df[(current_manager_df['id'] == id) &\
                                                        (current_manager_df['Месяц отгрузки'] == last_month)]
            denom_1 += shipment_in_completion_month['Сумма отгрузки'].sum()

            shipment_in_curr_month = current_manager_df[(current_manager_df['id'] == id) &\
                                                        (current_manager_df['Месяц отгрузки'] == last_month + 1)]
            numer_1 += shipment_in_curr_month['Сумма отгрузки'].sum()

        coeff1 = numer_1 / denom_1 if denom_1 > 0 else None

        denom_2 = 0
        numer_2 = 0

        for _, row in completed_ids.iterrows():
            id = row['id']
            last_month = row['Последний месяц реализации проекта']

            has_next_month = not current_manager_df[(current_manager_df['id'] == id) &\
                                                (current_manager_df['Месяц отгрузки'] == last_month + 1)].empty

            if not has_next_month:
                denom_2 += current_manager_df[(current_manager_df['id'] == id) &\
                                              (current_manager_df['Месяц отгрузки'] == last_month)]['Сумма отгрузки'].sum()

                numer_2 += current_manager_df[(current_manager_df['id'] == id) &\
                                              (current_manager_df['Месяц отгрузки'] == last_month + 2)]['Сумма отгрузки'].sum()

        coeff2 = numer_2 / denom_2 if denom_2 > 0 else None

        yearly_res.append({
            "Год": year,
            "Менеджер": manager,
            "Коэффициент 1 (год)": coeff1,
            "Коэффициент 2 (год)": coeff2
        })

yearly_results_df = pd.DataFrame(yearly_res)
```

### 4. Number of projects that didn't have shipments after completion, for each manager.

If a specific manager does not have such projects (i.e., all their projects had shipments after competion), then that manager will not be in the table

```python
last_month_for_each_project = merged_df[['id', 'AM', 'Последний месяц реализации проекта']].drop_duplicates()

def no_shipments_after_project_closing(id, last_month):
    return merged_df[(merged_df['id'] == id) & (merged_df['Месяц отгрузки'] > last_month)].empty

last_month_for_each_project['Нет отгрузки после завершения проекта'] = last_month_for_each_project.apply(
    lambda row: no_shipments_after_project_closing(row['id'], row['Последний месяц реализации проекта']), axis=1
)

projects_with_no_shipments_after_closing = last_month_for_each_project[~last_month_for_each_project['Нет отгрузки после завершения проекта']]
projects_with_no_shipments_after_closing_by_manager =  pd.DataFrame(projects_with_no_shipments_after_closing.groupby('AM')['id'].nunique())
projects_with_no_shipments_after_closing_by_manager = projects_with_no_shipments_after_closing_by_manager.rename(columns={
 'id': 'Кол-во проектов без отгрузки после завершения'
})
projects_with_no_shipments_after_closing_by_manager.index.set_names(['Менеджер'], inplace=True)

projects_with_no_shipments_after_closing_by_manager
```
